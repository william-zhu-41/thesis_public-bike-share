# Chicago data

```python
#import Chicago data (July 2019)
df = pd.read_csv('/content/drive/MyDrive/thesis/data/Divvy_Trips_2019_Q3.csv')
dfs = pd.read_csv('/content/drive/MyDrive/thesis/data/Divvy_Bicycle_Stations.csv')
```

```python
df['month'] = df['start_time'].apply(lambda x: x[:7])
df = df[df['month']=='2019-07']
df = df[['from_station_id', 'from_station_name', 'to_station_id', 'to_station_name']]
```

```python
dfs = dfs[['ID', 'Latitude',	'Longitude']]
dfout = df.merge(dfs, how='left', left_on = 'from_station_id', right_on='ID')
dfout = dfout.merge(dfs, how='left', left_on = 'to_station_id', right_on='ID', suffixes=('_start', '_end'))
```

```python
dfout = dfout.rename(columns={'from_station_name':'start_name', 
                        'Latitude_start':'start_lat',
                        'Longitude_start':'start_long', 
                        'to_station_name': 'end_name',
                        'Latitude_end': 'end_lat', 
                        'Longitude_end': 'end_long'})
dfout = dfout[['start_name', 'start_lat', 'start_long', 
               'end_name', 'end_lat', 'end_long']]
dfout = dfout.dropna(axis=0)    
```

```python
dfout = dfout[dfout['start_name']!=dfout['end_name']]
```

```python
dfout.to_csv('/content/drive/MyDrive/thesis/data/cleaned/chicago_201907.csv')
```

# Pittsburgh

```python
df = pd.read_excel('/content/drive/MyDrive/thesis/data/Healthy RIde Rentals 2019 Q3.xlsx')
dfs = pd.read_excel('/content/drive/MyDrive/thesis/data/Healthy Ride Station Locations Q3 2019.xlsx')
```

```python
df['month'] = df['Starttime'].apply(lambda x: str(x)[:7])
df = df[df['month']=='2019-07']
df = df[['From station id', 'From station name', 'To station id', 'To station name']]
df = df.dropna(axis=0)
df['From station id'] = df['From station id'].astype(int)
df['To station id'] = df['To station id'].astype(int)
```

```python
dfs = dfs[['Station #', 'Latitude',	'Longitude']]
dfout = df.merge(dfs, how='left', left_on = 'From station id', right_on='Station #')
dfout = dfout.merge(dfs, how='left', left_on = 'To station id', right_on='Station #', suffixes=('_start', '_end'))
```

```python
dfout = dfout.rename(columns={'From station name':'start_name', 
                        'Latitude_start':'start_lat',
                        'Longitude_start':'start_long', 
                        'To station name': 'end_name',
                        'Latitude_end': 'end_lat', 
                        'Longitude_end': 'end_long'})
dfout = dfout[['start_name', 'start_lat', 'start_long', 
               'end_name', 'end_lat', 'end_long']]
dfout = dfout.dropna(axis=0)
```

```python
dfout = dfout[dfout['start_name']!=dfout['end_name']]
```

```python
dfout = dfout[(dfout['start_lat']>0.0) & (dfout['end_lat']>0.0)]
```

```python
dfout.to_csv('/content/drive/MyDrive/thesis/data/cleaned/pittsburgh_201907.csv')
```
